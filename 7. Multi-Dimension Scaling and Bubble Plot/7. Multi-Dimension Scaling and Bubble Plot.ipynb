```python
import pandas as pd
import numpy as np
```

```python
csv_url = "https://simplemaps.com/static/data/country-cities/in/in.csv"
city_data = pd.read_csv(csv_url)
city_data.head()
```

```python
city_data.shape
```

```python
import geopy
from geopy.distance import geodesic
```

```python
def calculate_distance(city1, city2):
    return geodesic((city1["lat"], city1["lng"]), (city2["lat"], city2["lng"])).kilometers
```

```python
distanceMatrix = city_data.apply(lambda city1: city_data.apply(lambda city2: calculate_distance(city1, city2), axis=1), axis=1)
```

```python
distanceMatrix
```

```python
zones = {
    "North": ['Delhi', 'Himāchal Pradesh', 'Punjab', 'Uttar Pradesh', 'Haryāna', 'Jammu and Kashmīr', 'Chandīgarh'],
    "East": ['Bihār', 'Odisha', 'Jhārkhand', 'West Bengal'],
    "West": ['Rājasthān', 'Gujarāt', 'Goa', 'Mahārāshtra'],
    "South": ['Andhra Pradesh', 'Telangāna', 'Karnātaka', 'Kerala', 'Tamil Nādu', 'Puducherry'],
    "Central": ['Madhya Pradesh', 'Chhattīsgarh'],
    "North East": ['Assam']
}
```

```python
def assign_zone(city_name):
    for zone, cities in zones.items():
        if city_name in cities:
            return zone
    return "Unknown"
```

```python
city_data['zone'] = city_data['admin_name'].apply(assign_zone)
```

```python
city_data
```

```python
from sklearn.manifold import MDS
```

```python
mds = MDS(random_state=0, dissimilarity='precomputed', n_components=2, normalized_stress='auto')
fit_data = mds.fit_transform(distanceMatrix)
```

```python
fit_data
```

```python
import plotly.express as px
```

```python
city_data['x'] = fit_data[:, 0]
city_data['y'] = fit_data[:, 1]
city_data
```

```python
fig = px.scatter(city_data, x='x', y='y', size='population', color='zone', hover_name='city',
                 labels={'x': 'x', 'y': 'y', 'population': 'Population'},
                 title='Plotted using Plotly')
fig.show()
```
