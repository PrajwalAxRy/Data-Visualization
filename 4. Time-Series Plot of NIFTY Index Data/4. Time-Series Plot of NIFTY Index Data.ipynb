```python
import pandas as pd
import numpy as np
import nsepython
import plotly.express as px
```

```python
startDate = "2014-05-01"
endDate = "2023-09-10"
```

```python
indexFund = ["NIFTY 50", "NIFTY BANK", "NIFTY IT"]
```

```python
nifty50_data = nsepython.index_total_returns(indexFund[0], startDate, endDate)
niftyBank_data = nsepython.index_total_returns(indexFund[1], startDate, endDate)
niftyIT_data = nsepython.index_total_returns(indexFund[2], startDate, endDate)
```

```python
nifty50_data
```

```python
df_nifty50 = pd.DataFrame({'Date': nifty50_data['Date'], 'Value': pd.to_numeric(nifty50_data['TotalReturnsIndex'])})
df_niftybank = pd.DataFrame({'Date': niftyBank_data['Date'], 'Value': pd.to_numeric(niftyBank_data['TotalReturnsIndex'])})
df_niftyit = pd.DataFrame({'Date': niftyIT_data['Date'], 'Value': pd.to_numeric(niftyIT_data['TotalReturnsIndex'])})
```

```python
df_niftyit
```

```python
# Reverse the order of rows
df_nifty50 = df_nifty50[::-1].reset_index(drop=True)
df_niftybank = df_niftybank[::-1].reset_index(drop=True)
df_niftyit = df_niftyit[::-1].reset_index(drop=True)
```

```python
df_nifty50
```

```python
df_nifty50['Returns %'] = ((df_nifty50['Value'] - df_nifty50['Value'].iloc[0]) / df_nifty50['Value'].iloc[0])*100
df_niftybank['Returns %'] = ((df_niftybank['Value'] - df_niftybank['Value'].iloc[0]) / df_niftybank['Value'].iloc[0])*100
df_niftyit['Returns %'] = ((df_niftyit['Value'] - df_niftyit['Value'].iloc[0]) / df_niftyit['Value'].iloc[0])*100
```

```python
df_niftybank
```

```python
df_final = pd.DataFrame({'Date' : df_nifty50['Date'], 
                         'NIFTY 50' : df_nifty50['Returns %'], 
                         'NIFTY BANK' : df_niftybank['Returns %'],
                        'NIFTY IT' : df_niftyit['Returns %']})
```

```python
df_final
```

```python
import plotly.express as px

# Create a line plot
fig = px.line(df_final,
             x = 'Date',
             y = ['NIFTY 50', 'NIFTY BANK', 'NIFTY IT'],
             title = 'Percent Change in Index funds since May 2014')

fig.update_layout(
    xaxis_title = 'Date',
    yaxis_title = 'Value',
    legend_title_text = 'Index Funds',
    hovermode = 'x'
)

fig.show()
```
